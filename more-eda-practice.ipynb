# Third Task: More Graphing
I would love to go just a little further and introduce another kind of bar graph. You'll need seaborn (`pip install seaborn`). The `hue` parameter chooses the bars for each category. Of course, you're welcome to do as many as you want. Visual inspection can be critical in noticing trends that we do not see in the descriptive statistics.

Try the same plots with different variables. Be sure to update the axis labels and title. A general rule is that your viewer should not have to ask questions or know the data to understand your graph (assuming a reasonably smart viewer, of course) (and yes, I'm terrible at this, I admit). Also, feel free to pick nicer colors and other formatting options.



```python
import seaborn as sns
# catplot graph
sns.set_style('whitegrid')
sns.set_palette('husl',3)
sns.barplot(x='Gender',
            y='Income', 
            hue='Occupation',
            data=df,
           )
plt.legend(
    ncol=3, 
#    loc='upper center', 
    fontsize=10, 
    frameon=True, 
    edgecolor='black', 
    facecolor='lightgray',
    columnspacing=1.5
)
plt.title('Average Income, Broken down by Gender and Profession')
plt.xlabel('Gender')
plt.ylabel('Average Income (in local currency)')
```
